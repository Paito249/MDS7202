````python
def exploratory_data_analysis(dataframe):
  x,y = dataframe.shape
  name_col = list(dataframe.columns)
  first = dataframe.head()
  last = dataframe.tail()
  sample = dataframe.sample(n=5)
  dataframe[['math score','reading score','writing score']] = dataframe[['math score','reading score','writing score']].apply(pd.to_numeric)
  descrp = dataframe.describe()
  nulos = dataframe.isna()
  total_null = nulos.sum()
  unicos = dataframe.nunique()
  duplicados = dataframe.duplicated(subset='names')
  cant_dupl = duplicados.sum()
  filas_dup = dataframe[dataframe.duplicated(subset='names')]
  print(f'1.- El DataFrame tiene {x} filas y {y} columnas' )
  print(f'2.- El DataFrame esta copuesto por las siguientes columnas: {name_col}')
  print('3.- Ejemplos de filas del DataFrame:')
  print('Primeras 5 filas:')
  display(first)
  print('√öltimas 5 filas:')
  display(last)
  print('Muestreo aleatorio de 5 filas:')
  display(sample)
  print('4.- Descripci√≥n num√©rica del Dataframe:')
  display(descrp)
  print('5.- Cantidad de valores nulos por columna:')
  display(total_null)
  print('6.- Cantidad de valores √∫nicos por columna:')
  display(unicos)
  print(f'7.- Presenta un total de {cant_dupl} filas con nombres duplicados.')
  print('8.- Las filas con nombres duplicados son:')
  display(filas_dup)
  fig = px.box(dataframe, y=["math score", "reading score", "writing score"],
              labels={"variable": "Asignatura", "value": "Puntuaci√≥n"})
  fig.update_layout(title="Boxplot de Notas")
  fig.show()

````

**Justificaci√≥n:**


Seg√∫n la exploraci√≥n, y de acuerdo a las columnas del dataframe, se puede concluir que las columnas duplicadas por nombre (247) son un problema, ya que deber√≠a haber solo 1 registro por alumno. Por otro lado al ver los valores √∫nicos de nombres (1153) siguen siendo una gran cantidad en realci√≥n a los duplicados. Del boxplot se puede observar que el score de matem√°ticas es el que tiene el promedio m√°s bajo. Adem√°s, presenta valor identificados como outliers iguales a 0, lo que no ser√≠a un dato extra√±o considerando que son calificaciones.

### 1.3.1 Transformaci√≥n de notas a "Nota Chilena" [0.5 Puntos]

En una de las reuniones con Don Caguayo este les comenta su intenci√≥n de pasar todas sus notas a la "Escala Chilena". Para esto les facilita la siguiente ecuaci√≥n:

$$Nota \, Chilena= \% \, de \, logro*0.06 + 1$$

Utilizando la ecuaci√≥n entregada, cree una funci√≥n que le permita pasar a la escala Chilena las notas del dataframe y entregue las notas redondeadas al segundo decimal.

**To-Do**

- [x] Tranforme los valores a la escala chilena.

**Respuesta:**

````python
def nota_chilena(dataframe_in, columna):
  df=dataframe_in.copy()
  for i in range(len(df)):
    df[columna][i] = round(df[columna][i]*0.06+1, 2)
  return df

for materia in {'math score','reading score','writing score'}:
  df_grades = nota_chilena(df_grades,materia)
df_grades
````

### 1.3.3 Reprobados y Destacados [1.5 Punto, 0.5 Cada DataFrame]

Contentos con su desempe√±o, la Universidad les solicita que obtengan los alumnos quienes han reprobado el a√±o y los que han tenido desempe√±o de destacado. Para esto se considerar√° que un alumno ha reprobado el a√±o cuando el promedio de notas es inferior a 4.0 en la Escala Chilena, y se considerar√° destacado si posee notas superiores a 5.8 en todas las asignaturas (o sea notas superiores a 5.8 en `math score`, `reading score` y `writing score`).

Cree una funci√≥n que permita obtener tres Dataframes:
- El primero deber√° poseer solamente los alumnos que han reprobado en la prestigiosa universidad.
- El segundo debe entregar un DataFrame similar al original, pero marcado con una "R" el promedio de notas de los alumnos reprobados.
- El tercero ser√° un Data Frame con los alumnos destacados ordenados de mayor a menor.

**To-Do:**

- [x] Obtener DataFrame con los alumnos que reprobaron el a√±o.
- [x] Obtener DataFrame similar al original, pero marcados con una "R" el atributo GPA de los alumnos reprobados.
- [x] Obtener DataFrame con los alumnos destacados ordenados de mayor a menor.

**Respuesta:**

````python
def reprobados_destacados(dataframe_in):
  df=dataframe_in.copy()
  df_reprobados = df[df['GPA']<4].reset_index(drop = True)
  df.loc[df['GPA'] < 4.00, 'GPA'] = 'R'
  df_destacados=df[(df['math score'] > 5.8) & (df['reading score'] > 5.8) & (df['writing score'] > 5.8)]
  df_destacados = df_destacados.sort_values(by = "GPA", ascending = False)
  return df_reprobados,df,df_destacados
````

### 1.3.5 Quintiles de Rendimiento [0.5 Puntos]

El rector, intrigado por saber si existe una relaci√≥n entre el nivel educacional de los padres y las notas de sus hijos les plantea estudiar dicha relaci√≥n. Para esto, les solicita que desde el `DataFrame` que posee a todos los alumnos (`DataFrame` obtenido tras la limpieza), generen una nueva columna que clasifique en quintiles de rendimiento para sus alumnos: `['Baj√≠simo', 'Bajo', 'Medio', 'Alto', 'Muy alto']`. (investigar `qcut()`)

Hecho esto, grafique a trav√©s de un scatter plot el nivel educacional de los padres v/s el promedio de los alumnos, utilice los cuantiles generados anteriormente para mostrar en que cuantil se encuentra cada uno de los datos.

Realizado el grafico, responda la siguiente pregunta: **¬øEs posible visualizar alguna tendencia en el gr√°fico?.**



**Respuesta:**

````python
# C√≥digo para quintiles
# La variable que contenga los quintiles debe llamarse GPA cuantiles para que el gr√°fico funcione.

df_grades['GPA cuantiles'] = pd.qcut(df_grades['GPA'], q=5, labels=["Baj√≠simo", "Bajo", "Medio","Alto","Muy Alto"])
# C√≥digo de gr√°fico
fig = px.scatter(
    df_grades.sort_values(by='GPA'),
    x='GPA',
    y="parental level of education",
    hover_data=["gender"],
    color="GPA cuantiles",
    title="GPA por Nivel de educaci√≥n de los padres",
    color_discrete_sequence=["tomato", "orange", "yellow", "greenyellow", "green"]
)

fig
````

A partir del scatterplot, es evidente la relaci√≥n entre el nivel educacional de los padres y el GPA de los estudiantes. Se puede ver que en el nivel educacional de padres m√°s alto, el GPA m√°s bajo es 3.68, siendo adem√°s, el √∫nico caso que reprobar√≠a en este nivel. Por el contrario, en los niveles educacionales de padres m√°s bajos, se observan los menores promedios de estudiantes, adem√°s de una alta concentraci√≥n de promedios al rededor del 4.

Considerando ahora la distribuci√≥n de cuartiles, el rendimiento "Baj√≠simos" se nota co mayor presencia en los niveles educacionales de padres m√°s bajos. Por otro lado, el rendimiento "Muy Alto", se encuentra principalmente concentrado en los niveles educacionales de padres m√°s altos.

A modo de conclusi√≥n, s√≠ se pueden identificar tendencias: A mayor nivel educacional de padres, mejor es el rendimiento (menos estudiantes con rendimiento "Baj√≠simo").
